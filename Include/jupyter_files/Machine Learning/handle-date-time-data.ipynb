```python
import pandas as pd

df=pd.read_csv("D:\powerBI\PowerBI Tutorial\Orders.csv")

# df['Date']=pd.to_datetime(df['Date'])


df['Year']=pd.to_numeric(df['Date'].str[-5:-1])
df['Month']=pd.to_numeric(df['Date'].str[-8:-6])
df['Day']=pd.to_numeric(df['Date'].str[-11:-9])
df['New_date']=pd.to_datetime(df[['Year','Month','Day']])

df['M_Year']=df['New_date'].dt.year
df['M_Month']=df['New_date'].dt.month

df['M_Day']=df['New_date'].dt.day
# df['M_Week']=df['New_date'].dt.week

df['M_Weekday']=df['New_date'].dt.weekday
df['M_Quarter']=df['New_date'].dt.quarter

df['M_Dayofyear']=df['New_date'].dt.dayofyear
df['M_Dayofweek']=df['New_date'].dt.dayofweek


df
```
