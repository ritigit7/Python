```python
import matplotlib.pyplot as plt
from scipy import stats
import numpy as np

x =sorted(np.random.randint(0,100,50))
y =sorted(np.random.randint(0,100,50))
print(x,y)

slope, intercept, r, p, std_err = stats.linregress(x, y)
print(slope, intercept, "\nreletion:",r, p, std_err)

def myfunc(i):
  return slope * i + intercept

print(myfunc(15))

mymodel = list(map(myfunc, x))

plt.scatter(x, y)
plt.plot(x, y)
plt.plot(x, mymodel)
plt.show()
```
